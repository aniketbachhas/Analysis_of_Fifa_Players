```python
# importing Libraries 
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Load the CSV file and check
fifaData = pd.read_csv('/content/players_20.csv')
fifaData.shape
```

```python
fifaData.head()
```

```python
# Print all Columns in this DataSet
for col in fifaData.columns:  
    print(col)
```

```python
#Height vs Weight Analysis
sns.scatterplot(x=fifaData['weight_kg'], y=fifaData['height_cm'])
```

```python
fifaData['nationality'].value_counts()
```

```python
# Extracting Names of Country
firstFiveCountriesNames = fifaData['nationality'].value_counts()[0:5].keys()    
list(firstFiveCountriesNames)
```

```python
firstFiveCountriesPlayers = fifaData['nationality'].value_counts()[0:5]
# Extracting Total no.of Players from these Country
firstFiveCountriesPlayers
```

```python
plt.figure(figsize=(8,5))
plt.bar(firstFiveCountriesNames,firstFiveCountriesPlayers,color='skyblue')
plt.title('TOP FIVE COUNTRIES')
plt.grid(True)
plt.show()
```

```python
# Creating Bar chart to show the wages of top 5 player
playerSalaryData = fifaData[["short_name",'wage_eur']]
PSD = playerSalaryData.sort_values(by=['wage_eur'],ascending= False)
PSD.head()
```

```python
playersName = PSD['short_name'].head(5)  # creating variable for Players Name
playersSalary = PSD['wage_eur'].head(5)  # creating variable for Players Salaries
```

```python
plt.figure(figsize=(8,5))
plt.bar(playersName,playersSalary,color=['b','g','r','pink','orange'])
plt.title('TOP FIVE PLAYER SALARIES')
plt.ylabel("SALARIES")
plt.xlabel('PLAYER NAMES')
plt.grid(True)
plt.show()
```
